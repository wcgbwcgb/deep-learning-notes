```python
from torch.utils.data import DataLoader
from torch import nn
from torch.nn import Conv2d
import torchvision

dataset = torchvision.datasets.CIFAR10("./CIFAR10_dataset", train=False, transform=torchvision.transforms.ToTensor(),
                                       download=True)
dataloader = DataLoader(dataset, batch_size=64, drop_last=True)

## conv2d layer
class MyModel(nn.Module):
    def __init__(self):
        super().__init__() 
        self.conv1 = Conv2d(in_channels=3, out_channels=6, kernel_size=3, stride=1, padding=0)
    
    def forward(self, x):
        output = self.conv1(x)
        return output
    
model = MyModel()
print(model)
```

```python
import torch
from torch.nn import MaxPool2d

# Pooling Layer
input = torch.tensor([[1,2,0,3,1],
                      [0,1,2,3,1],
                      [1,2,1,0,0],
                      [5,2,3,1,1],
                      [2,1,0,1,1]])

input = torch.reshape(input, (-1, 1, 5, 5))

class PoolingModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.maxpool1 = MaxPool2d(kernel_size=3, ceil_mode=True)

    def forward(self, input):
        output = self.maxpool1(input)
        return output
    
poolmodel = PoolingModel()
output = poolmodel(input)
print(output)
```

```python
from torch.nn import ReLU

# nonlinear activation: relu
input = torch.tensor([[1,-0.5],
                      [-1,3]])

input = torch.reshape(input, (-1, 1, 2, 2))
print(input.shape)

class ReLUModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.relu1 = ReLU()

    def forward(self, input):
        output = self.relu1(input)
        return output
    
relumodel = ReLUModel()
output = relumodel(input)
print(output)
```

```python
from torch.nn import Linear

# Linear Layer

class LinearModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear1 = Linear(196608, 10)

    def forward(self, input):
        output = self.linear1(input)
        return output

linearmodel = LinearModel()

for data in dataloader:
    imgs, targets = data
    print(imgs.shape)

    output = torch.flatten(imgs)

    print(output.shape)
    output = linearmodel(output)
    print(output.shape)
```
