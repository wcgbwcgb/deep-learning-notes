```python
from torchvision import transforms
from PIL import Image

img_path = "../learn_pytorch/dataset/train/ants_image/0013035.jpg"
img = Image.open(img_path)
print(img)

# How to use transforms
tensor_trans = transforms.ToTensor()
tensor_img = tensor_trans(img)
print(tensor_img)
```

```python
import cv2
cv_img = cv2.imread(img_path) # numpy array
print(type(cv_img))
```

```python
from torch.utils.tensorboard import SummaryWriter

writer = SummaryWriter("logs")
writer.add_image("tensor_img", tensor_img)
writer.close()
```

### Introducing classes in transforms

```python
writer = SummaryWriter("logs")

# ToTensor
# tensor_trans = transforms.ToTensor()
# tensor_img = tensor_trans(img)
# print(tensor_img)

# Normalize
print(tensor_img[0][0][0])
trans_norm = transforms.Normalize([6, 0.5, 3],[2, 1, 5.5])
img_norm = trans_norm(tensor_img)
print(img_norm[0][0][0])
# writer.add_image("normalized_image", img_norm, 2)

# Resize
print(img.size)
trans_resize = transforms.Resize((512, 512))
img_resize = trans_resize(img)  # return resized image
img_resize = tensor_trans(img_resize)  # convert into tensor
# writer.add_image("resize", img_resize)
print(img_resize)

# Compose - resize - 2
trans_resize2 = transforms.Resize(256)
trans_compose = transforms.Compose([trans_resize2, tensor_trans])
img_resize_2 = trans_compose(img)
print("img2:", img_resize_2.size)
# writer.add_image("compose_resize", img_resize_2, 3)

# RandomCrop
trans_random = transforms.RandomCrop(256)
trans_compose2 = transforms.Compose([trans_random, tensor_trans])
for i in range(10):
    img_crop = trans_compose2(img)
    writer.add_image("RandomCrop", img_crop, i)
```
