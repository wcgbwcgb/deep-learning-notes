```python
import torchvision
from torch import nn

vgg16_false = torchvision.models.vgg16(pretrained=False)
vgg16_true = torchvision.models.vgg16(pretrained=True)

print(vgg16_false)
vgg16_false.classifier.add_module("add_linear", nn.Linear(1000, 10)) # add module
print(vgg16_false)

print(vgg16_true)
vgg16_true.classifier[6] = nn.Linear(4096, 10) # change module
print(vgg16_true)
```
