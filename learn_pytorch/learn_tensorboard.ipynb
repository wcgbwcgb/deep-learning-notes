```python
from torch.utils.tensorboard import SummaryWriter
writer = SummaryWriter("logs")

for i in range(100):
    y = 2*i
    writer.add_scalar("y=2x", y, i)  # tag: title, scalar_value: y-axis, global_step = x-axis

writer.close()

# to open the event file, we need to input "tensorboard --logdir=logs --port=6006" in the terminal
```
