```python
import torch
import torchvision

vgg16 = torchvision.models.vgg16(pretrained=False)

# saving method 1
torch.save(vgg16, "vgg16_method1.pth") # structure + parameters

# loading method 1
torch.load("vgg16_method1.pth")

# saving method 2
torch.save(vgg16.state_dict(), "vgg16_method2.pth") # only parameters

# loading method 1
vgg16_load = torchvision.models.vgg16(pretrained=False)
vgg16_load.load_state_dict(torch.load("vgg16_method2.pth"))
```
